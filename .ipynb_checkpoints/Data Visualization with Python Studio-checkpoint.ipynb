```python
# Here is the setup of the dataframe. Feel free to use .info() or .head() to get a better understanding of what is inside the dataframe!

import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline

df = pd.read_csv("books.csv", error_bad_lines=False)

years = []

for y in df['publication_date']:
    date = y.split("/")
    years.append(date[2])

df["year"] = years
```

```python
df.info()
```

With your dataframe set up, time to start creating visualizations!

```python
df.head()
```

```python
# Use your Python skills to create a visualization here.
Author= df[df.authors =='J.K. Rowling']
Author
```

```python
Totalbooks = Author['title'].value_counts().sum()
Totalbooks
```

```python
plt.plot(Author.year, Author.title)

plt.title('Books published by J.K. Rawling', color = 'black')
plt.xlabel('Years',color ='black') 
plt.ylabel('Title', color = 'black')
plt.xticks(color = 'black')
plt.yticks(color = 'black')
plt.show()
```

```python
plt.style.use("ggplot")
plt.figure(figsize=(8,6))
sb.regplot(x = Author["ratings_count"], y = Author["text_reviews_count"])
plt.title('Ratings Vs reviews of J.K. Rawling', color = 'black')
plt.show()
```
